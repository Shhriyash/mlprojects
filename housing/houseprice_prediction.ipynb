```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('housing.csv')
```

```python
data
```

```python
data.info()
```

```python
#deleting null values
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data['ocean_proximity'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#dropping target variable 
X=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
#target variable dropped from dataset and splitted for testing
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.67,random_state=42)
```

```python
#joining training datasets
trainingdata = X_train.join(y_train)
```

```python
trainingdata.ocean_proximity.value_counts()
```

```python
#one hot encoding 
pd.get_dummies(trainingdata.ocean_proximity).astype(int)
```

```python
#changing their datatype and joining back to the training dataset
trainingdata=trainingdata.join(pd.get_dummies(trainingdata.ocean_proximity).astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
trainingdata
```

```python
#importing liner regression model
from sklearn.linear_model import LinearRegression
```

```python
#dropping target variable from training dataset
X_train,y_train=trainingdata.drop(['median_house_value'],axis=1),trainingdata['median_house_value']
```

```python
#instantiate the model
mymodel=LinearRegression()
```

```python
X_train
```

```python
#fitting into a model
mymodel.fit(X_train,y_train)
```

```python
#feeding  test data
testingdata = X_test.join(y_test)


```

```python
X_test
```

```python
testingdata=testingdata.join(pd.get_dummies(testingdata.ocean_proximity).astype(int)).drop(['ocean_proximity'],axis=1) 
```

```python
testingdata
```

```python
X_test,y_test=testingdata.drop(['median_house_value'],axis=1),testingdata['median_house_value']
```

```python
X_test
```

```python
from sklearn.metrics import mean_squared_error as mse

y_pred = mymodel.predict(X_test)
```

```python
mse_test = mse(y_test, y_pred)
```

```python
import math

math.sqrt(mse_test)
```

```python
mymodel.score(X_test,y_test)
```
